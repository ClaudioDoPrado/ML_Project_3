```python
from sklearn.datasets import  fetch_california_housing
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor,AdaBoostRegressor, GradientBoostingRegressor

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv(r"C:\Users\gebruiker\IH_LESSONS\Projects\ML_Project_3\calories.csv")
df.head()
```

```python
# Set pandas future option to avoid the warning
pd.set_option('future.no_silent_downcasting', True)

# Proceed with your replacement
df['Gender'] = df['Gender'].replace({
    'male': 0,
    'female': 1
})

# Check the result
df.head()
```

```python
features = df.drop(columns = ["User_ID","Calories"])
target= df["Calories"]
```

```python
# Drop non-numeric columns, like 'User_ID' and 'Gender', before correlation
X_numeric = features.select_dtypes(include=[float, int])

# Now, compute correlation matrix
corr = X_numeric.corr()

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Among Numeric Features')
plt.show()
```

```python
# 3. Feature Importance using Random Forest Regressor
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_train_scaled.head()
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
X_test_scaled.head()
```

```python
##Bagging
bagging_reg = BaggingRegressor(DecisionTreeRegressor(max_depth=6), n_estimators=100, max_samples = 1000)
bagging_reg.fit(X_train_scaled, y_train)
```

```python
pred = bagging_reg.predict(X_test_scaled)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", bagging_reg.score(X_test_scaled, y_test))
```

```python
sns.regplot(x=pred,y=y_test)
```

```python
##AdaBoost
ada_reg = AdaBoostRegressor(DecisionTreeRegressor(max_depth=20),n_estimators=100)
ada_reg.fit(X_train_scaled, y_train)
```

```python
pred = ada_reg.predict(X_test_scaled)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", ada_reg.score(X_test_scaled, y_test))
```

```python
##RandomForest
rf_model = RandomForestRegressor(n_estimators=100, random_state=20)
rf_model.fit(X_train_scaled, y_train)
```

```python
pred = rf_model.predict(X_test_scaled)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", rf_model.score(X_test_scaled, y_test))
```
